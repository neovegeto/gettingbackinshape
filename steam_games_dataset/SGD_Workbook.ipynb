```python
#Correlation Matrix:

# To compute the correlation matrix for numerical columns:
correlation_matrix = df.corr()
display(correlation_matrix)
```

```python
# Create the Heatmap:
# Use the seaborn library to create a heatmap from the correlation matrix. 
# You can customize the appearance of the heatmap using various parameters:
# https://seaborn.pydata.org/generated/seaborn.heatmap.html

# Set the style of the heatmap
sns.set(style="white")

# Create a heatmap using the correlation matrix
plt.figure(figsize=(10, 8))  # Adjust the figure size if needed
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python


# Select only the numeric columns for plotting histograms
numeric_columns = df.select_dtypes(include=['number'])

# Set the figure size for the histograms
plt.figure(figsize=(12, 8))  # Adjust the figure size if needed

# Plot histograms for numeric columns
for column in numeric_columns.columns:
    plt.hist(numeric_columns[column], bins=20, alpha=0.7, label=column)

# Add labels, legend, and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histograms for Numeric Columns')
plt.legend()

# Display the histograms
plt.show()
```

```python
# Select only the numeric columns for plotting histograms
numeric_columns = df.select_dtypes(include=['number'])

# Set the figure size
plt.figure(figsize=(12, 8))

# Create histograms for each numeric column
for column in numeric_columns.columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins=20)  # Remove the alpha parameter
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram for {column}')
    plt.show()
```

```python
# Set the Seaborn style (optional)
sns.set(style="whitegrid")

# Filter out non-numeric columns
numeric_columns = df.select_dtypes(include=['number'])

# Create box plots for each numeric column
for column in numeric_columns.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df[column])
    plt.xlabel('Column')
    plt.ylabel('Value')
    plt.title(f'Box Plot for Column: {column}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Select numeric columns for density plots
numeric_columns = df.select_dtypes(include=['number'])

# Set the Seaborn style
sns.set(style="whitegrid")

# Create density plots for selected numeric columns
for column in numeric_columns.columns:
    plt.figure(figsize=(8, 6))
    sns.kdeplot(df[column], shade=True)  # Create density plot
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.title(f'Density Plot for {column}')
    plt.show()
```

```python
# Select numeric columns for scatter plots
numeric_columns = df.select_dtypes(include=['number'])

# Set the figure size
plt.figure(figsize=(10, 8))

# Create scatter plots for selected pairs of numeric columns
scatter_pairs = [
    ('Peak CCU', 'Metacritic score'),
    ('User score', 'Positive'),
    ('User score', 'Negative'),
    ('Average playtime forever', 'Median playtime forever')
]

for x_column, y_column in scatter_pairs:
    plt.figure(figsize=(8, 6))
    plt.scatter(df[x_column], df[y_column], alpha=0.5)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.title(f'Scatter Plot: {x_column} vs. {y_column}')
    plt.grid(True)
    plt.show()
```

```python
# Convert 'Release date' column to datetime
df['Release date'] = pd.to_datetime(df['Release date'])

# Create a time series plot for release dates
plt.figure(figsize=(12, 8))
plt.plot(df['Release date'], marker='o')
plt.ylabel('Release Date')
plt.xlabel('Count')
plt.title('Game Releases Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create a histogram for estimated owners
plt.figure(figsize=(12, 8))
plt.hist(df['Estimated owners'], bins=20, alpha=0.7)
plt.xlabel('Estimated Owners')
plt.ylabel('Frequency')
plt.title('Distribution of Estimated Owners')
plt.tight_layout()
plt.show()
```
